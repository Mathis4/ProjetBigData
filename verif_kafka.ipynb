```python
from kafka import KafkaConsumer
import json

# Variables de configuration
KAFKA_TOPIC = "github_repos"
KAFKA_BOOTSTRAP_SERVERS = "localhost:9092"

# Initialisation du consommateur Kafka
consumer = KafkaConsumer(
    KAFKA_TOPIC,
    bootstrap_servers=KAFKA_BOOTSTRAP_SERVERS,
    group_id="github_repos_consumer_group",  # Groupe de consommateurs
    value_deserializer=lambda x: json.loads(x.decode('utf-8'))  # D√©s√©rialisation du message en JSON
)

# Fonction pour afficher les donn√©es des d√©p√¥ts
def display_repos_from_kafka():
    print("En attente de nouveaux messages sur Kafka...")
    for message in consumer:
        repo_data = message.value  # Contenu du message Kafka (les donn√©es du d√©p√¥t)
        print(f"Nom du d√©p√¥t: {repo_data['repo_name']}")
        print(f"URL du d√©p√¥t: {repo_data['html_url']}")
        print(f"Description: {repo_data.get('description', 'Pas de description')}")
        print(f"Date de cr√©ation: {repo_data['created_at']}")
        print("-" * 50)

if __name__ == "__main__":
    try:
        display_repos_from_kafka()  # Consommer et afficher les messages
    except KeyboardInterrupt:
        print("Arr√™t du consommateur Kafka.")
    finally:
        consumer.close()
```
